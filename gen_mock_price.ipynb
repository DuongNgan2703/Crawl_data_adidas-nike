```python
import json
import numpy as np
# Path to your pickle file
data_main_path = 'C:/Users/nvatu/data_crawl_com_main.json'

# Loading the data from the pickle file
with open(data_main_path, 'r') as file:
    all_data = json.load(file)
len(all_data)
```

```python
for item in all_data:
    if 'price' in item and isinstance(item['price'], str):
        # Remove the dollar sign from the price
        item['price'] = item['price'].replace('$', '')

        try:
            item['price'] = float(item['price'])
        except ValueError:
            print(f"Could not convert price value '{item['price']}' to float.")
```

```python
def get_mock_prices(price):
  mean = price
  std = ((0.3 - 0.1) * np.random.rand() + 0.1) * mean
  num = np.random.randint(3, 6, 1)
  mock_prices = np.random.normal(loc=mean,
                                scale=std,
                                size=num)
  return mock_prices.tolist()
```

```python
list_mock_price = []
for i in all_data:
    mock_prices = []
    mock_prices = get_mock_prices(i['price'])
    list_mock_price.append(mock_prices) 
list_mock_price
```

```python
import json

# Tên của hai tệp JSON bạn muốn gộp
file1_name = "C:/Users/nvatu/data_craw_com_v.json"
file2_name = "C:/Users/nvatu/data_crawl_com.json"

# Đọc dữ liệu từ file1 với mã hóa UTF-8
with open(file1_name, 'r', encoding='utf-8') as file1:
    data1 = json.load(file1)

# Đọc dữ liệu từ file2 với mã hóa UTF-8
with open(file2_name, 'r', encoding='utf-8') as file2:
    data2 = json.load(file2)

# Gộp danh sách từ cả hai tệp vào một danh sách mới
merged_data = data1 + data2

# Tên của tệp mới
merged_file_name = "data_crawl_com_main.json"

# Ghi dữ liệu gộp vào tệp mới với mã hóa UTF-8
with open(merged_file_name, 'w', encoding='utf-8') as merged_file:
    json.dump(merged_data, merged_file, indent=4)

print(f"Gộp dữ liệu từ {file1_name} và {file2_name} thành {merged_file_name}")
```
